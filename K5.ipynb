```r
train = read.csv("train3.csv")
```

```r
train
```

```r
library(caTools)

set.seed(88)

spl = sample.split(train, SplitRatio = 0.7)

Train = subset(train, spl==TRUE)

Test = subset(train, spl==FALSE)
```

```r
summary(Train)
```

```r
Female_Train = subset(Train, Train$Gender == "Female")
```

```r
Female_Test = subset(Test, Test$Gender == "Female")
```

```r
Female_Train_18 = subset(Female_Train, Female_Train$AgeGroup == "18-24")
```

```r
Female_Test_18 = subset(Female_Test, Female_Test$AgeGroup == "18-24")
```

```r
plot(Female_Train_18$EducationLevel, Female_Train_18$Income)
```

```r
library(randomForest)

EForest = randomForest(Party ~ .-USER_ID, data = Female_Train_18, nodesize = 40, ntree = 200)
```

```r
PredictForest = predict(EForest, newdata = Female_Test_18)
```

```r
table(Female_Test_18$Party, PredictForest)
```

```r
ETree = rpart(Party ~ .-USER_ID ,data = Female_Train_18, method="class", minbucket=40)
```

```r
PredictCART = predict(ETree, newdata = Female_Test_18, type = "class")
```
