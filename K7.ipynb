```r
set.seed(144)

spl = sample.split(train, SplitRatio  = 0.7)

Train = subset(train, spl == TRUE)

Test = subset(train, spl == FALSE)
```

```r
limitedTrain = Train
limitedTrain$X = NULL
limitedTrain$Party = NULL
limitedTrain$USER_ID = NULL

limitedTest = Test
```

```r
limitedTest$X = NULL
limitedTest$USER_ID = NULL
limitedTest$Party = NULL
```

```r
limitedTest
```

```r
kosDist = dist(limitedTrain, method="euclidean")

kosHierClust = hclust(kosDist, method="ward.D")
```

```r
plot(kosHierClust)
```

```r
kosDist_Test = dist(limitedTest, method="euclidean")

kosHierClust_Test = hclust(kosDist_Test, method="ward.D")
```

```r
plot(kosHierClust_Test)
```

```r
hierGroups = cutree(kosHierClust, k = 5)
```

```r
HierCluster1 = subset(Train, hierGroups == 1)

HierCluster2 = subset(Train, hierGroups == 2)

HierCluster3 = subset(Train, hierGroups == 3)

HierCluster4 = subset(Train, hierGroups == 4)

HierCluster5 = subset(Train, hierGroups == 5)

```

```r
str(HierCluster5)
```

```r
HierCluster_Test1 = subset(Test, hierGroups_Test == 1)

```

```r
HierCluster_Test2 = subset(Test, hierGroups_Test == 2)

HierCluster_Test3 = subset(Test, hierGroups_Test == 3)

HierCluster_Test4 = subset(Test, hierGroups_Test == 4)

HierCluster_Test5 = subset(Test, hierGroups_Test == 5)

```

# K 均值

```r
set.seed(144)

km = kmeans(limitedTrain, centers = 5)

```

```r
library(flexclust)

km.kcca = as.kcca(km, limitedTrain)

clusterTrain = predict(km.kcca)

clusterTest = predict(km.kcca, newdata=limitedTest)
```

```r
table(clusterTest)
```

```r
Train$X = NULL
Test$X = NULL
Train$USER_ID = NULL
```

```r
stocksTrain1 = subset(Train, clusterTrain == 1)

stocksTrain2 = subset(Train, clusterTrain == 2)

stocksTrain3 = subset(Train, clusterTrain == 3)

stocksTrain4 = subset(Train, clusterTrain == 4)

stocksTrain5 = subset(Train, clusterTrain == 5)




stocksTest1 = subset(Test, clusterTest == 1)

stocksTest2 = subset(Test, clusterTest == 2)

stocksTest3 = subset(Test, clusterTest == 3)

stocksTest4 = subset(Test, clusterTest == 4)

stocksTest5 = subset(Test, clusterTest == 5)
```

```r
StocksModel1 = glm(Party ~ ., data=stocksTrain1, family=binomial)

StocksModel2 = glm(Party ~ ., data=stocksTrain2, family=binomial)

StocksModel3 = glm(Party ~ ., data=stocksTrain3, family=binomial)

StocksModel4 = glm(Party ~ ., data=stocksTrain4, family=binomial)

StocksModel5 = glm(Party ~ ., data=stocksTrain5, family=binomial)
```

```r
StocksModel1 = glm(Party ~ .-USER_ID, data=stocksTrain1, family=binomial)

StocksModel2 = glm(Party ~ .-USER_ID, data=stocksTrain2, family=binomial)

StocksModel3 = glm(Party ~ .-USER_ID, data=stocksTrain3, family=binomial)

StocksModel4 = glm(Party ~ .-USER_ID, data=stocksTrain4, family=binomial)

StocksModel5 = glm(Party ~ .-USER_ID, data=stocksTrain5, family=binomial)
```

```r
PredictTest1 = predict(StocksModel1, newdata = stocksTest1, type="response")

PredictTest2 = predict(StocksModel2, newdata = stocksTest2, type="response")

PredictTest3 = predict(StocksModel3, newdata = stocksTest3, type="response")

PredictTest4 = predict(StocksModel4, newdata = stocksTest4, type="response")

PredictTest5 = predict(StocksModel5, newdata = stocksTest5, type="response")
```
